# Vclay Calculation

Using Larianov (1969) Method, taking an example from a legacy well in the offshore, Gulf of Mexico

```python
# Import important library
import lasio
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Example from MP 299 Block - UWI 17-724-40687-00

```python
# Read the LAS file
las_file = '1IN DUAL INDUCTION FOCUSED LOG GAMMA RAY_WL163075.las'
las = lasio.read(las_file)

# Convert to a DataFrame
df = las.df()

# Reset the index to get DEPT as a column
df.reset_index(inplace=True)

# Calculate statistics for GR
gr_mean = df['GR'].mean()
gr_median = df['GR'].median()
gr_10th_percentile = np.percentile(df['GR'].dropna(), 10)
gr_90th_percentile = np.percentile(df['GR'].dropna(), 90)
gr_min = df['GR'].min()
gr_max = df['GR'].max()

# Plotting GR without shading
fig, ax = plt.subplots(figsize=(3, 10))

# Set up the plot axes
ax1 = ax

ax10 = ax1.twiny()
ax10.xaxis.set_visible(False)

ax1.plot(df["GR"], df["DEPT"], color="green", linewidth=0.5)
ax1.set_xlabel("GR")
ax1.xaxis.label.set_color("green")
ax1.set_xlim(df["GR"].min(), df["GR"].max())
ax1.set_ylabel("Depth (ft)")
ax1.tick_params(axis='x', colors="green")
ax1.spines["top"].set_edgecolor("green")
ax1.title.set_color('green')

ax1.set_ylim(df["DEPT"].max(), df["DEPT"].min())
ax1.grid(which='major', color='lightgrey', linestyle='-')
ax1.xaxis.set_ticks_position("top")
ax1.xaxis.set_label_position("top")
ax1.spines["top"].set_position(("axes", 1.02))

plt.tight_layout()
fig.subplots_adjust(wspace=0.15)
plt.show()

# Print statistics for GR
print(f"GR Mean: {gr_mean:.2f}")
print(f"GR Median: {gr_median:.2f}")
print(f"GR 10th Percentile: {gr_10th_percentile:.2f}")
print(f"GR 90th Percentile: {gr_90th_percentile:.2f}")
print(f"GR Min: {gr_min:.2f}")
print(f"GR Max: {gr_max:.2f}")
```

## GR value distribution throughout the well

```python
# Create a histogram for GR
plt.figure(figsize=(10, 6))
plt.hist(df['GR'].dropna(), bins=50, edgecolor='black', alpha=0.7)

# Add lines for mean, median, 10th percentile, and 90th percentile
plt.axvline(gr_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {gr_mean:.2f}')
plt.axvline(gr_median, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {gr_median:.2f}')
plt.axvline(gr_10th_percentile, color='green', linestyle='dashed', linewidth=2, label=f'10th Percentile: {gr_10th_percentile:.2f}')
plt.axvline(gr_90th_percentile, color='purple', linestyle='dashed', linewidth=2, label=f'90th Percentile: {gr_90th_percentile:.2f}')

# Add labels and title
plt.xlabel('GR Value')
plt.ylabel('Frequency')
plt.title('Histogram of GR Values')
plt.legend()

# Show the plot
plt.show()
```

## Shade GR track based on mud baseline
Yellow color is sand-dominated, gray color is mud-dominated

```python
# Plotting GR with shading
fig, ax = plt.subplots(figsize=(3, 10))

# Set up the plot axes
ax1 = ax

ax10 = ax1.twiny()
ax10.xaxis.set_visible(False)

ax1.plot(df["GR"], df["DEPT"], color="green", linewidth=0.5)
ax1.set_xlabel("GR")
ax1.xaxis.label.set_color("green")
ax1.set_xlim(df["GR"].min(), df["GR"].max())
ax1.set_ylabel("Depth (ft)")
ax1.tick_params(axis='x', colors="green")
ax1.spines["top"].set_edgecolor("green")
ax1.title.set_color('green')

# Shade the interval where GR is below the median value as yellow
ax1.fill_betweenx(df["DEPT"], df["GR"], where=(df["GR"] < gr_median), color='yellow', alpha=0.5)

# Shade the interval where GR is above the median value as grey
ax1.fill_betweenx(df["DEPT"], df["GR"], where=(df["GR"] >= gr_median), color='grey', alpha=0.5)

ax1.set_ylim(df["DEPT"].max(), df["DEPT"].min())
ax1.grid(which='major', color='lightgrey', linestyle='-')
ax1.xaxis.set_ticks_position("top")
ax1.xaxis.set_label_position("top")
ax1.spines["top"].set_position(("axes", 1.02))

plt.tight_layout()
fig.subplots_adjust(wspace=0.15)
plt.show()
```

## Converting GR to Vclay (Larionov, 1969)
Plotted in log format

```python
# Calculate Vclay using GR min and max
# Calculate Gamma Ray Index (GRI)
GR_min = df['GR'].min()
GR_max = df['GR'].max()

df['GRI'] = (df['GR'] - GR_min) / (GR_max - GR_min)

# Calculation for Vclay using the Larionov (1969) young rock method
df['Vclay'] = 0.083 * (2**(3.7 * df['GRI']) - 1.0)
df['Vclay'] = df['Vclay'].clip(lower=0, upper=1)

# Plotting Vclay
fig, ax = plt.subplots(figsize=(3, 10))

# Set up the plot axes
ax1 = ax

ax10 = ax1.twiny()
ax10.xaxis.set_visible(False)

ax1.plot(df["Vclay"], df["DEPT"], color="blue", linewidth=0.5)
ax1.set_xlabel("Vclay")
ax1.xaxis.label.set_color("blue")
ax1.set_xlim(0, 1)
ax1.set_ylabel("Depth (ft)")
ax1.tick_params(axis='x', colors="blue")
ax1.spines["top"].set_edgecolor("blue")
ax1.title.set_color('blue')

ax1.set_ylim(df["DEPT"].max(), df["DEPT"].min())
ax1.grid(which='major', color='lightgrey', linestyle='-')
ax1.xaxis.set_ticks_position("top")
ax1.xaxis.set_label_position("top")
ax1.spines["top"].set_position(("axes", 1.02))

plt.tight_layout()
fig.subplots_adjust(wspace=0.15)
plt.show()
```

## Shade Vclay track based on Vclay content
Yellow color is Vclay <0.3, orange is 0.3 < Vclay < 0.6, grey color is Vclay >0.6

```python
# Filter data for the interval of interest: e.g. all interval (3800 ft - 8600 ft)
df_interval = df[(df['DEPT'] >= 3800) & (df['DEPT'] <= 8600)]

# Plotting Vclay
fig, ax = plt.subplots(figsize=(3, 10))

# Set up the plot axes
ax1 = ax

ax10 = ax1.twiny()
ax10.xaxis.set_visible(False)

# Vclay track with color shading
for i in range(len(df_interval) - 1):
    depth_top = df_interval.iloc[i]['DEPT']
    depth_bottom = df_interval.iloc[i + 1]['DEPT']
    vclay_top = df_interval.iloc[i]['Vclay']
    vclay_bottom = df_interval.iloc[i + 1]['Vclay']
    
    color = 'yellow' if vclay_top < 0.3 else 'khaki' if vclay_top < 0.6 else 'whitesmoke'
    ax1.fill_betweenx([depth_top, depth_bottom], 0, [vclay_top, vclay_bottom], color=color)

ax1.plot(df_interval["Vclay"], df_interval["DEPT"], color="black", linewidth=0.5)
ax1.set_xlabel("Vclay")
ax1.xaxis.label.set_color("blue")
ax1.set_xlim(0, 1)
ax1.set_ylabel("Depth (ft)")
ax1.tick_params(axis='x', colors="blue")
ax1.spines["top"].set_edgecolor("blue")
ax1.title.set_color('blue')

ax1.set_ylim(8600, 3800)
ax1.grid(which='major', color='lightgrey', linestyle='-')
ax1.xaxis.set_ticks_position("top")
ax1.xaxis.set_label_position("top")
ax1.spines["top"].set_position(("axes", 1.02))

plt.tight_layout()
fig.subplots_adjust(wspace=0.15)
plt.show()
```

```python
# Filter data for the interval of interest: e.g. all interval (3800 ft - 8600 ft)
df_interval_avg = df[(df['DEPT'] >= 3800) & (df['DEPT'] <= 8600)]

# Calculate the average Vclay for the interval
average_vclay = df_interval_avg['Vclay'].mean()

# Output the result
print(f"Average Vclay (clay fraction): {average_vclay:.2f}")
```
